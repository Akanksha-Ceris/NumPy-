```python
# Univariate - Numerical
data = {
    "salary":[23400,5600,3000,4000,5600,2600,8500,8000,22456,9089]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
df.head(5)
```

```python
df.shape
```

```python
# Line Plot:
plt.plot(df["salary"],color = "red",marker = "o",linestyle = "--",linewidth = "2")
plt.show()
```

```python
# Histogram
plt.hist(df["salary"],bins = 5,color = 'green')
plt.show()
```

```python
# Boxplot:
plt.boxplot(df["salary"])
plt.show()
```

```python
df.head()
```

```python
df.info()
```

```python
# Pie chart
count = df["dept"].value_counts()
count
```

```python
plt.pie(count)
plt.show()
```

```python
plt.pie(count,labels= count.index,autopct="%1.2f", explode =[0] * count)
plt.show()
```

```python
# countplot
plt.bar(count.index,count,color = ["green","black","pink","red"])
plt.show()
```

```python
# Bivariant -numerical -numerical
df.head()
```

```python
df["Age"] = [34,56,78,90,12,34,56,54,4576,98,90]
df.head()
```

```python
# Scater plot:
plt.scatter(df["Age"],df["salary"],color=["orange"])
plt.show()
```

```python
# Line plot:
plt.plot(df["Age"],df["salary"],color = "red",marker = "",linewidth = "2")
plt.grid()
plt.show()
```

```python
# Bar chart
sort_age = df.sort_values("Age")

plt.bar(sort_age["Age"],df["salary"],color = "green",align="center")
plt.show()
```

```python
# Bivarate : Numerical  and categorical
```

```python
hr_sal=df[df["dept"]== "HR"] ["salary"]
hr_sal
```

```python
it_sal =df[df["dept"]=="IT"]["salary"]
it_sal
```

```python
fi_sal= df[df["dept"] == "FINANCE"]["salary"]
fi_sal
```

```python
#Boxplot
```

```python
plt.boxplot([hr_sal,it_sal,fi_sal])
plt.show()
```

```python
# Pie chart
salary_by_dept = df.groupby("dept")["salary"].sum()
salary_by_dept
```

```python
# PLOT it in pie chart
plt.pie(salary_by_dept,labels=salary_by_dept.index,autopct = "%1.2f")
plt.axis("equal")
plt.show()
```

```python
# bar  plot
# I am going to show you mean values of all of my dapartment salary
hr_mean=sum(hr_sal)/len(hr_sal)
it_mean=sum(it_sal)/len(it_sal)
fi_mean=sum(fi_sal)/len(fi_sal)
```

```python
plt.bar(["HR","IT","FINANCE"],[hr_mean,it_mean,fi_mean],color =["green","black","red"])
plt.grid()
plt.show()
```

```python
# Multivariant Analysis: 3 numerical columns:
df.head()
```

```python
df["experience"] = [1,2,3,4,5,6,7,9,8,6.5,3.4]
df.head()
```

```python
# plot a pie chart
plt.pie(salary_by_dept,labels=salary_by_dept.index,autopct="%1.2f",shadow=True,explode=[0.1,0,2.1])
plt.axis("equal")
plt.show()
```

```python
# Bubble PLOT:
plt.scatter(df["Age"],df["salary"],s = df["experience"]*20,color="blue")
plt.title("Age vs  salary")
plt.title("Age")
plt.title("salary")
plt.show()
```

```python
print(df.columns)
```

```python
# 2 numerical and 1 categorical column:
```

```python
plt.scatter(df["Age"],df["salary"],c=df["dept"].map({"HR":"Yellow","IT":"blue","FINANCE":"orange"}),label=df["dept"])
plt.xlabel("Age")
plt.ylabel("salary")
plt.title("Age vs salary vs Dept")
plt.legend()
plt.show()

```

```python
color = {"HR":"yellow","IT":"blue","FINANCE":"orange"}
for dept,color in color.items():
  df_dept = df[df["dept"]==dept]
  plt.scatter(df_dept["Age"],df_dept["salary"],c = color,label = dept)
plt.legend()

```
