# **Brodcasting**

The term broadcasting describe  how numpy treats arrays with different shapes during arithmetic operations.
The smaller arraay is "broadcast" across the larger so that they have compatible  shape


```python
# same shape
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)
print(a)
print(b)
```

```python
print(a +b)
```

```python
# diff shpae
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)
print(a)
print(b)
print(a +b)
```

Numpy unfuncs
what are unfuncs:
unfuncs stads for universal functions and they are numpy functions that operate on the ndarray object


```python
# without unfunc , we can use python's built-in zip() method:
x = [1,2,3,4]
y = [4,5,6,7]
z = []
for i , j in zip(x,y):
  z.append(i + j)

print(z)
```

```python
# numpy has a func for this , called add(x,y ) that will produce the same result
# with unfunc , we can use the add() function
import numpy as np
x = [1,2,3,4]
y = [4,5,6,7]
z = np.add(x,y)
print(z)
```
